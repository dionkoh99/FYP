```python
#Exploratory analysis - Movie description
import nltk
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize, sent_tokenize 
stop_words = set(stopwords.words('english')) 
from string import punctuation
import collections
import re
```

```python
# Creating a list of words from all the movie descriptions in df data
all_desc = []
for i in range(0,df.shape[0]):
    desc = df.loc[i,'description']
    desc1 = desc.lower()
    # Removing numbers, *** and www links from the data
    desc2 = re.sub('[0-9]+\S+|\s\d+\s|\w+[0-9]+|\w+[\*]+.*|\s[\*]+\s|www\.[^\s]+','',desc1)
    # Removing punctuation
    for p in punctuation:
        desc2 = desc2.replace(p,'')
    all_desc.append(desc2)
```

```python
# Creating word tokens for all the descriptions
final_list = []
for desc in all_desc:
    word_list = word_tokenize(desc)
    final_list.extend(word_list)
```

```python
# 3. Tagging parts of speech
pos_tagged = nltk.pos_tag(final_list)

# 4. Identifying the most common parts of speech
tag_fd = nltk.FreqDist(tag for (word, tag) in pos_tagged)
tag_fd.most_common()[:5]
```

```python
# Removing stopwords
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english')) 

list_wo_stopwords = []
for w in final_list:
    if w not in stop_words:
        list_wo_stopwords.append(w)
        
# 3. Tagging parts of speech
pos_tagged_wo_sw = nltk.pos_tag(list_wo_stopwords)

# 4. Identifying the most common parts of speech
tag_fd_wo_sw = nltk.FreqDist(tag for (word, tag) in pos_tagged_wo_sw)
tag_fd_wo_sw.most_common()[:5]
```

```python
from nltk.stem import WordNetLemmatizer
import matplotlib.pyplot as plt
lemmatizer = WordNetLemmatizer()

# Lemmatization without specifying parts of speech
list_lemmatized = []
for word in list_wo_stopwords:
    list_lemmatized.append(lemmatizer.lemmatize(word))

word_freq_lem = dict(collections.Counter(list_lemmatized))
keys = list(word_freq_lem.keys())
values = list(word_freq_lem.values())
df_lem = pd.DataFrame({'words':keys,'freq':values})
display(df_lem.sort_values(by = 'freq',ascending = False)[:10])

from wordcloud import WordCloud
from collections import Counter
word_could_dict=Counter(word_freq_lem)
wordcloud = WordCloud(width = 1000, height = 500).generate_from_frequencies(word_could_dict)

plt.figure(figsize=(15,8))
plt.imshow(wordcloud)
```

# Text data cleaning (Movie description)

```python
#Initial cleaning function (remove symbols, numbers, punctuation and stopwords)
import nltk
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize, sent_tokenize 
stop_words = set(stopwords.words('english')) 
from string import punctuation
import collections
import re
import string

#List unwanted words
unwanted_words = stopwords.words('english') + list(string.punctuation)
def textcleaning (sentence):
    sentence1 = sentence.lower() #lower case
    sentence2 = re.sub(r'[^A-Za-z\s]','',sentence1)#remove all non-alpha character
    tokens = word_tokenize(sentence2) #tokenize the sentence
    filtered_sentence = [w for w in tokens if not w in unwanted_words] #filter the unwanted words
    return " ".join(filtered_sentence) #join them back into sentence


data2['c_description'] = data2['description'].apply(lambda x: textcleaning(x))
    
```

```python
# Lemmatize the words
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.corpus import wordnet
lemmatizer = nltk.stem.WordNetLemmatizer()
wordnet_lemmatizer = WordNetLemmatizer()

#return the postag for the token
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return None

#
def lemmatize_sentence(sentence):
    #tokenize the sentence and find the POS tag for each token
    nltk_tagged = nltk.pos_tag(nltk.word_tokenize(sentence))
    #tuple of (token, wordnet_tag)
    wordnet_tagged = map(lambda x: (x[0], get_wordnet_pos(x[1])), nltk_tagged)
    lemmatized_sentence = []
    for word, tag in wordnet_tagged:
        if tag is None:
            #if there is no available tag, append the token as is
            lemmatized_sentence.append(word)
        else:
            #else use the tag to lemmatize the token
            lemmatized_sentence.append(lemmatizer.lemmatize(word, tag))
    return " ".join(lemmatized_sentence)



# Lemmatizing
data2['c_description'] = data2['c_description'].apply(lambda x: lemmatize_sentence(x))
    
    
```

# Data modelling


```python
train_df = data2.filter(['c_description','log_trans_gross_income'])
display(train_df)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import Ridge, SGDRegressor, LinearRegression
import numpy as np
from sklearn import metrics
import xgboost as xgb
    
vectorizer = TfidfVectorizer(min_df = 10)
x = vectorizer.fit_transform(train_df['c_description'])
y = train_df['log_trans_gross_income']

#Split into test, train
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)

#Linear regression
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)
y_pred = linear_reg.predict(X_test)


#Ridge regression
ridge_reg = Ridge()
ridge_params = {'alpha':[20,10,5,4,3,2,1,0.2,0.01]}
ridge_gs = GridSearchCV(estimator = ridge_reg, param_grid = ridge_params, scoring='neg_mean_squared_error', cv = 5)
ridge_gs.fit(X_train, y_train)
print('best ridge model:',ridge_gs.best_estimator_)
print('best ridge model score (nMSE): ', ridge_gs.best_score_)
best_ridge_model = ridge_gs.best_estimator_
pred = best_ridge_model.predict(X_test)


#SGDRegressor
sgd_reg = SGDRegressor()
sgd_params = {'penalty':['none','l2','l1'],'alpha':[1e-4, 2e-4, 5e-4, 1e-3, 2e-3, 5e-3, 1e-2, 2e-2, 5e-2, 0.01]}
sgd_gs = GridSearchCV(estimator = sgd_reg, param_grid = sgd_params, scoring = 'neg_mean_squared_error', cv=5)
sgd_gs.fit(X_train, y_train)
print('best sgd model:',sgd_gs.best_estimator_)
print('best sgd model score (nMSE): ', sgd_gs.best_score_)
best_sgd_model = sgd_gs.best_estimator_
pred2 = best_sgd_model.predict(X_test)

#XGBoost 
xgb_reg = xgb.XGBRegressor()
xgb_params = {'nthread':[4], 
              'objective':['reg:linear'],
              'learning_rate': [.03, 0.05, .07], 
              'max_depth': [5, 6, 7],
              'min_child_weight': [4],
              'silent': [1],
              'subsample': [0.7],
              'colsample_bytree': [0.7],
              'n_estimators': [1000]}

xgb_gs = GridSearchCV(xgb_reg,xgb_params,cv = 2,n_jobs = 5,scoring = 'neg_mean_squared_error')
xgb_gs.fit(X_train, y_train)
print('best xgb model:',xgb_gs.best_estimator_)
print('best xgb model score (nMSE): ', xgb_gs.best_score_)
best_xgb_model = xgb_gs.best_estimator_
pred3 = best_xgb_model.predict(X_test)


#Create a dataframe for actual figure and predicted figure.(LinearRegression)
pred_df4 = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
#Convert back to normal figure.
pred_df4['Actual'] = np.expm1(pred_df4['Actual'])
pred_df4['Predicted'] = np.expm1(pred_df4['Predicted'])
pred_df4['difference'] = abs(pred_df4['Actual'] - pred_df4['Predicted'])

#Create a dataframe for actual figure and predicted figure.(RidgeRegressor)
pred_df = pd.DataFrame({'Actual': y_test, 'Predicted': pred})
#Convert back to normal figure.
pred_df['Actual'] = np.expm1(pred_df['Actual'])
pred_df['Predicted'] = np.expm1(pred_df['Predicted'])
pred_df['difference'] = abs(pred_df['Actual'] - pred_df['Predicted'])

#Create a dataframe for actual figure and predicted figure.(SGDRegressor)
pred_df2 = pd.DataFrame({'Actual': y_test, 'Predicted': pred2})
#Convert back to normal figure.
pred_df2['Actual'] = np.expm1(pred_df2['Actual'])
pred_df2['Predicted'] = np.expm1(pred_df2['Predicted'])
pred_df2['difference'] = abs(pred_df2['Actual'] - pred_df2['Predicted'])

#Create a dataframe for actual figure and predicted figure.(XGBoost)
pred_df3 = pd.DataFrame({'Actual': y_test, 'Predicted': pred3})
#Convert back to normal figure.
pred_df3['Actual'] = np.expm1(pred_df3['Actual'])
pred_df3['Predicted'] = np.expm1(pred_df3['Predicted'])
pred_df3['difference'] = abs(pred_df3['Actual'] - pred_df3['Predicted'])

#Evaluation metrics
print('\n Linear Regression: \n')
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('Scaled MAE (USD):', pred_df4['difference'].mean())
print('RMSPE(%): ', (np.sqrt(np.mean(np.square((y_test - y_pred) / y_test)))) * 100)
print('MAPE(%): ', np.mean(np.abs((y_test - y_pred) / y_test)) * 100)

print('\n RidgeRegressor: \n')
print('MAE:', metrics.mean_absolute_error(y_test, pred))
print('MSE:', metrics.mean_squared_error(y_test, pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, pred)))
print('Scaled MAE(USD):', pred_df['difference'].mean())
print('RMSPE(%): ', (np.sqrt(np.mean(np.square((y_test - pred) / y_test)))) * 100)
print('MAPE(%): ', np.mean(np.abs((y_test - pred) / y_test)) * 100)

print('\n SGDRegressor: \n')
print('MAE:', metrics.mean_absolute_error(y_test, pred2))
print('MSE:', metrics.mean_squared_error(y_test, pred2))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, pred2)))
print('Scaled MAE(USD):', pred_df2['difference'].mean())
print('RMSPE(%): ', (np.sqrt(np.mean(np.square((y_test - pred2) / y_test)))) * 100)
print('MAPE(%): ', np.mean(np.abs((y_test - pred2) / y_test)) * 100)

print('\n XGBRegressor: \n')
print('MAE:', metrics.mean_absolute_error(y_test, pred3))
print('MSE:', metrics.mean_squared_error(y_test, pred3))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, pred3)))
print('Scaled MAE(USD):', pred_df3['difference'].mean())
print('RMSPE(%): ', (np.sqrt(np.mean(np.square((y_test - pred3) / y_test)))) * 100)
print('MAPE(%): ', np.mean(np.abs((y_test - pred3) / y_test)) * 100)
```
